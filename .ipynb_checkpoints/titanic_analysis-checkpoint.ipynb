```python
import pandas as pd

# Membaca file CSV
df = pd.read_csv('titanic.csv')  # Ganti 'data.csv' dengan path file yang sesuai

# Mengubah kolom 'sex' menjadi biner (misalnya 'Laki-laki' jadi 1, 'Perempuan' jadi 0)
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})

# Mengganti nilai object di kolom Age menjadi biner
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

# Melakukan cek kolom Age yang memiliki nilai Null
print(df['Age'].isnull().sum())

# Simpan data frame ke dalam csv
df.to_csv('data_filled.csv', index=False)
```

```python
# Data setelah transformasi

import pandas as pd

df = pd.read_csv('data_filled.csv')

df.head()
```

```python
"""
Analisis berdasarkan umur dan jenis kelamin
"""

# 1. Hitung jumlah yang survive
ts = df['Survived'].value_counts()[1]
print("Jumlah penumpang yang selamat: ", ts)

# 2. Hitung jumlah yang survive berdasarkan Sex
ts_by_sex = df.groupby('Sex')['Survived'].sum()
print("Total jenis kelamin penumpang yang selamat: ", ts_by_sex)

# 3. Hitung jumlah yang survive berdasarkan umur dan jenis kelamin
survived_passangers = df[df['Survived'] == 1]

age_counts = survived_passangers['Age'].value_counts().sort_index()
print(age_counts)
age_gender_counts = survived_passangers.groupby(['Sex', 'Age']).size().unstack(fill_value=0)
print(age_gender_counts)

# 4. Crosstab
survival_analysis = pd.crosstab([df['Sex'], df['Age']], df['Survived'])
print(survival_analysis)
#survival_analysis.to_csv('survival_analysis.csv', index=False)
```

```python
"""
Analisis berdasarkan kelas dan 
"""

# 5. Hitung jumlah survived berdasarkan Kelas
ts_class = pd.crosstab(df['Pclass'], df['Survived'])
print(ts_class)

# 6. Analisis jenis kelamin yang mendominasi kelas
ts_class_sex = df.groupby(['Sex', 'Pclass'])['Survived'].sum().reset_index()
print(ts_class_sex)

# 7. Analisis umur yang menempati kelas 1
first_class_passengers = df[df['Pclass'] == 1]

age_gender_counts = first_class_passengers.groupby(['Age', 'Sex']).size().reset_index(name='Counts')

most_common = age_gender_counts.loc[age_gender_counts['Counts'].idxmax()]

most_common_age = most_common['Age']
most_common_sex = most_common['Sex']
most_common_count = most_common['Counts']

print(f"Umur yang paling banyak menempati kelas 1: {most_common_age}, Jenis Kelamin: {most_common_sex} dengan jumlah: {most_common_count}")
```

```python
"""
Visualisasi Data
"""
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('data_filled.csv')

# 1. Menampilkan grafik data selamat dan tidak selamat

ts = df['Survived'].value_counts()
print("Jumlah penumpang : ", ts)

plt.bar(['Tidak Selamat', 'Selamat'], ts, color=['red','green'])
plt.title("Total penumpang selama dan tidak selamat di kapal Titanic")
plt.xlabel('Status Penumpang')
plt.ylabel('Jumlah Penumpang')

plt.show()
```

```python
# 2. Hitung Survived beserta jenis kelaminnya
ts_by_sex = df.groupby('Sex')['Survived'].sum()

ts_by_sex.plot(kind='bar', color=['pink','blue'])
print("Jumlah Penumpang Selamat: ", ts_by_sex)
plt.title('Jumlah Penumpang Selamat Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah Penumpang Selamat')
plt.show()
```

```python
# 2. Hitung Survived beserta jenis kelaminnya
ts_by_age = df.groupby('Age')['Survived'].sum()

max_age = ts_by_age.idxmax()

data_filtered = ts_by_age[[max_age]]

data_filtered.plot(kind='bar', color=['blue', 'red'], figsize=(8,5))
plt.title('Jumlah Penumpang Selamat Berdasarkan Umur Max dan Min')
plt.xlabel('Umur')
plt.ylabel('Jumlah Penumpang Selamat')

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
# Menghitung jumlah penumpang yang selamat berdasarkan jenis kelamin dan kelas
ts_class_sex = df.groupby(['Sex', 'Age'])['Survived'].sum().reset_index()

# Membuat pivot table untuk mempermudah plotting
pivot_table = ts_class_sex.pivot(index='Age', columns='Sex', values='Survived')

# Membuat stacked bar chart untuk menampilkan analisis berdasarkan kelas dan jenis kelamin
pivot_table.plot(kind='bar', stacked=True, color=['pink', 'blue'], figsize=(10,8))

# Menambahkan judul dan label
plt.title('Jumlah Penumpang Selamat Berdasarkan Jenis Kelamin dan Umur')
plt.xlabel('Kelas Penumpang')
plt.ylabel('Jumlah Penumpang Selamat')

# Menambahkan legenda
plt.legend(title='Jenis Kelamin', bbox_to_anchor=(1.05, 1), loc='upper left')

# Menampilkan grafik
plt.tight_layout()
plt.show()
```

```python
# 2. Hitung Survived beserta jenis kelaminnya
ts_by_pclass = df.groupby('Pclass')['Survived'].sum()

ts_by_pclass.plot(kind='bar', color=['gold','silver', 'yellow'])
print("Jumlah Penumpang Selamat: ", ts_by_pclass)
plt.title('Jumlah Penumpang Selamat Berdasarkan Kelas')
plt.xlabel('Kelas')
plt.ylabel('Jumlah Penumpang Selamat')
plt.show()
```

```python
# Menghitung jumlah penumpang yang selamat berdasarkan jenis kelamin dan kelas
ts_class_sex = df.groupby(['Sex', 'Pclass'])['Survived'].sum().reset_index()

# Membuat pivot table untuk mempermudah plotting
pivot_table = ts_class_sex.pivot(index='Pclass', columns='Sex', values='Survived')

# Membuat stacked bar chart untuk menampilkan analisis berdasarkan kelas dan jenis kelamin
pivot_table.plot(kind='bar', stacked=True, color=['pink', 'blue'], figsize=(10,6))

# Menambahkan judul dan label
plt.title('Jumlah Penumpang Selamat Berdasarkan Jenis Kelamin dan Kelas')
plt.xlabel('Kelas Penumpang')
plt.ylabel('Jumlah Penumpang Selamat')

# Menambahkan legenda
plt.legend(title='Jenis Kelamin', bbox_to_anchor=(1.05, 1), loc='upper left')

# Menampilkan grafik
plt.tight_layout()
plt.show()

# Menampilkan hasil data
print(ts_class_sex)
```

```python
# Filter penumpang yang berada di kelas 1
first_class_passengers = df[df['Pclass'] == 1]

# Mengelompokkan data berdasarkan umur dan jenis kelamin, menghitung jumlahnya
age_gender_counts = first_class_passengers.groupby(['Age', 'Sex']).size().reset_index(name='Counts')

# Membuat bar chart untuk menampilkan jumlah penumpang berdasarkan umur dan jenis kelamin
age_gender_counts.pivot(index='Age', columns='Sex', values='Counts').plot(kind='bar', figsize=(10,6), color=['pink', 'blue'])

# Menambahkan judul dan label
plt.title('Distribusi Penumpang Kelas 1 Berdasarkan Umur dan Jenis Kelamin')
plt.xlabel('Umur')
plt.ylabel('Jumlah Penumpang')

# Menambahkan legenda
plt.legend(title='Jenis Kelamin', bbox_to_anchor=(1.05, 1), loc='upper left')

# Menampilkan grafik
plt.tight_layout()
plt.show()
```
